# Theory Questions:-

```python
#Q1. What is hypothesis testing in statistics.
#Ans. Hypothesis testing is the process of using data to assess whether a specific claim (hypothesis) about a population parameter is likely to be true or false.
```

```python
#Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis.
#Ans. The null hypothesis is a statement that there is no effect, no difference, or no relationship in the population.It represents the status quo or default assumption.
```

```python
#Q3. What is the significance level in hypothesis testing, and why is it important.
#Ans. The significance level, denoted by α (alpha), is the threshold probability used to decide whether to reject the null hypothesis (H₀).It represents the maximum risk you're willing to accept of making a Type I error — that is, rejecting H₀ when it is actually true.
```

```python
#Q4. What does a P-value represent in hypothesis testing.
#Ans. A p-value is the probability of observing a result at least as extreme as the one obtained in your sample, assuming the null hypothesis (H₀) is true.
```

```python
#Q5. How do you interpret the P-value in hypothesis testing.
#Ans. The p-value is the probability of obtaining a result as extreme or more extreme than the observed one, assuming the null hypothesis (H₀) is true.
```

```python
#Q6. What are Type 1 and Type 2 errors in hypothesis testing.
#Ans. In hypothesis testing, decisions are made based on sample data — and like any decision-making process, errors can occur.
```

```python
#Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing.
#Ans. In hypothesis testing, one-tailed and two-tailed tests determine where to look for evidence against the null hypothesis (H₀).
#* Two-Tailed Test:- A test where the alternative hypothesis looks for a difference in either direction.
#*  One-Tailed Test:- A test where the alternative hypothesis looks for a difference in only one direction — either greater than or less than.
```

```python
#Q8. What is the Z-test, and when is it used in hypothesis testing.
#Ans. The Z-test is a statistical test used to determine whether there is a significant difference between sample data and a population parameter when the population standard deviation is known.It uses the Z-score (standard score) to determine how far a sample mean is from the population mean in units of standard deviation.
```

```python
#Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing.
#Ans. A Z-score (standard score) tells you how many standard deviations a data point is from the mean of a distribution.
#> Z= X−μ/σ

```

```python
#Q10. What is the T-distribution, and when should it be used instead of the normal distribution.
#Ans. The T-distribution (or Student's t-distribution) is a probability distribution used in hypothesis testing and confidence intervals when the sample size is small and/or the population standard deviation is unknown.It looks like the normal distribution, but has heavier tails, meaning more probability in the extremes — which accounts for extra uncertainty in small samples.
                                                                                                                       
```

```python
#Q11. What is the difference between a Z-test and a T-test.
#Ans. A Z-test is a statistical test used to determine if there is a significant difference between sample and population means when the population standard deviation (σ) is known and the sample size is large (n ≥ 30).
#Ans. A T-test is used to determine if there is a significant difference between sample and population means when the population standard deviation is unknown and/or the sample size is small (n < 30).
```

```python
#Q12. What is the T-test, and how is it used in hypothesis testing.
#Ans. A T-test is a statistical test used to determine whether there is a significant difference between means — especially when the sample size is small and the population standard deviation is unknown.It helps evaluate whether the difference between sample results and expectations under the null hypothesis is due to random chance or a real effect.
#* Use a T-test when:
#> Population standard deviation is unknown
#> Sample size is small (typically n < 30)
#> Data is approximately normally distributed
#> You're comparing means
```

```python
#Q13. What is the relationship between Z-test and T-test in hypothesis testing.
#Ans. The Z-test and T-test are both statistical tools used in hypothesis testing to determine whether sample data significantly differ from a population (or another sample). They are closely related — both test means, but are used under different assumptions.
```

```python
#Q14. What is a confidence interval, and how is it used to interpret statistical results.
#Ans. A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like a mean or proportion) with a specified level of confidence.It gives an estimated range instead of a single point — helping quantify uncertainty.
```

```python
#Q15. What is the margin of error, and how does it affect the confidence interval.
#Ans. The margin of error (MoE) is the maximum amount that a sample estimate (like a mean or proportion) is expected to differ from the true population value due to random sampling variability.It represents the range above and below the sample statistic in a confidence interval.
```

```python
#Q16. How is Bayes' Theorem used in statistics, and what is its significance.
#Ans. Bayes’ Theorem is a fundamental rule in probability that allows you to update the probability of a hypothesis as more evidence or information becomes available.
#* Bayes’ Theorem is essential for updating probabilities in light of new data and is the foundation of Bayesian statistics.
```

```python
#Q17. What is the Chi-square distribution, and when is it used.
#Ans. The Chi-square (χ²) distribution is a probability distribution used primarily in categorical data analysis. It is the distribution of the sum of the squares of independent standard normal variables.
#> The Chi-square distribution is used in inferential statistics, especially for tests involving categorical variables.
```

```python
#Q18. What is the Chi-square goodness of fit test, and how is it applied.
#Ans. The Chi-square goodness of fit test is a statistical method used to determine whether a set of observed frequencies matches an expected distribution.It tests whether data follows a specific theoretical distribution (like uniform, binomial, Poisson, etc.).
#* Use this test when:
#. You have categorical data.
#> You want to see if your observed counts differ significantly from expected counts based on a known or hypothesized distribut
```

```python
#Q19. What is the F-distribution, and when is it used in hypothesis testing.
#Ans. The F-distribution is a continuous probability distribution used primarily to compare two variances or test multiple means in hypothesis testing.
```

```python
#Q20. What is an ANOVA test, and what are its assumptions.
#Ans. ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more independent groups to determine if at least one group mean is significantly different from the others.Instead of comparing each group pair individually (like multiple t-tests), ANOVA tests all group means simultaneously, reducing the chance of Type I error.
#* Assumptions of ANOVA::-
#1. Independence
#2. Normality
#3. Homogeneity of Variance (Equal variances)
```

```python
#Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
#Ans. 
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
        x (int): 0 or 1 (the outcome)
        p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
        float: Probability of observing outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")
    return p**x * (1 - p)**(1 - x)
```

```python
#Q9. Write a Python function to calculate Z-scores from a dataset and plot them.
#Ans. 
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores for a dataset and plot them.

    Parameters:
        data (list or array-like): The input numeric dataset.
    """
    # Convert to NumPy array
    data = np.array(data)
    
    # Compute mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Print a few Z-scores
    print("First 5 Z-scores:", z_scores[:5])

    # Plot original values vs Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(data, label='Original Data', marker='o', linestyle='--', alpha=0.7)
    plt.plot(z_scores, label='Z-scores', marker='x', linestyle='-', alpha=0.7)
    plt.title('Original Data vs. Z-scores')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    return z_scores
```

```python
#Q10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
#Ans. 
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a non-normal population (exponential distribution)
population = np.random.exponential(scale=2.0, size=100000)

# Step 2: Define function to draw samples and compute sample means
def sample_means(population, sample_size, num_samples):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size)
        means.append(np.mean(sample))
    return np.array(means)

# Step 3: Simulate for different sample sizes
sample_sizes = [5, 30, 100]
plt.figure(figsize=(15, 4))

for i, n in enumerate(sample_sizes):
    means = sample_means(population, sample_size=n, num_samples=1000)
    plt.subplot(1, 3, i + 1)
    plt.hist(means, bins=30, density=True, alpha=0.7, edgecolor='black')
    plt.title(f'Sample Size = {n}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)

plt.suptitle('Central Limit Theorem Demonstration (Exponential Distribution)', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
#Ans. 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculate the Z-score of a data point.
    
    Parameters:
        x (float): Data point
        mean (float): Mean of the dataset
        std (float): Standard deviation of the dataset
        
    Returns:
        float: Z-score
    """
    return (x - mean) / std

# Example data
mean = 100
std = 15
x_value = 130

# Step 1: Calculate Z-score
z = calculate_z_score(x_value, mean, std)
print(f"Z-score of x = {x_value} (mean = {mean}, std = {std}): {z:.2f}")

# Step 2: Plot standard normal distribution with the Z-score location
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=0, scale=1)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')
plt.fill_between(x, y, where=(x <= z), color='skyblue', alpha=0.5, label=f'Z ≤ {z:.2f}')
plt.axvline(z, color='red', linestyle='--', label=f'Z-score = {z:.2f}')
plt.title("Standard Normal Distribution with Z-score Marked")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Q15. Implement hypothesis testing using Z-statistics for a sample dataset.
#Ans. 
import numpy as np
from scipy.stats import norm

# Given data
population_mean = 100      # claimed population mean (μ)
sample_mean = 96           # observed sample mean (x̄)
sample_std = 10            # standard deviation (σ)
sample_size = 50           # sample size (n)
alpha = 0.05               # significance level

# Step 1: Calculate the standard error
standard_error = sample_std / np.sqrt(sample_size)

# Step 2: Calculate Z-score
z_score = (sample_mean - population_mean) / standard_error

# Step 3: Critical value for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Step 4: Decision
print(f"Z-score: {z_score:.2f}")
print(f"Z-critical (±): {z_critical:.2f}")

if abs(z_score) > z_critical:
    print("❌ Reject the null hypothesis: Mean is significantly different.")
else:
    print("✅ Fail to reject the null hypothesis: No significant difference.")
```

```python
#Q16.  Create a confidence interval for a dataset using Python and interpret the result.
#Ans. 
import numpy as np
from scipy.stats import norm

# Step 1: Create or load your sample data
data = [48, 52, 50, 49, 51, 53, 47, 50, 52, 49]  # Example data
confidence_level = 0.95

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # use ddof=1 for sample std deviation
n = len(data)

# Step 3: Calculate standard error
standard_error = sample_std / np.sqrt(n)

# Step 4: Get the Z-critical value
z_critical = norm.ppf((1 + confidence_level) / 2)

# Step 5: Calculate margin of error and confidence interval
margin_of_error = z_critical * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 6: Output results
print(f"Sample mean: {sample_mean:.2f}")
print(f"{int(confidence_level * 100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
#Ans. 
import numpy as np
from scipy.stats import norm

def z_test_sample_mean(sample_mean, population_mean, sample_std, n, alpha=0.05):
    # Step 1: Calculate standard error
    se = sample_std / np.sqrt(n)

    # Step 2: Calculate Z-score
    z = (sample_mean - population_mean) / se

    # Step 3: Critical Z-value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Step 4: Interpretation
    print(f"Sample Mean: {sample_mean}")
    print(f"Population Mean: {population_mean}")
    print(f"Z-score: {z:.4f}")
    print(f"Z-critical (±): {z_critical:.4f}")

    if abs(z) > z_critical:
        print("❌ Reject the null hypothesis (significant difference).")
    else:
        print("✅ Fail to reject the null hypothesis (no significant difference).")

# Example values
sample_mean = 96
population_mean = 100
sample_std = 10
n = 50

# Run the Z-test
z_test_sample_mean(sample_mean, population_mean, sample_std, n)
```

```python
#Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
#Ans. 
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean using t-distribution.
    
    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (e.g., 0.95 for 95%)
    
    Returns:
        (mean, lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    se = stats.sem(data)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n - 1)  # Two-tailed t-critical value

    margin_of_error = t_critical * se
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print("\n📘 Interpretation:")
    print(f"We are {int(confidence * 100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")

    return mean, lower, upper

# 🔍 Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=30)
confidence_interval(sample_data, confidence=0.95)
```

```python
#Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data
#Ans.
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean.
    
    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (default 0.95 for 95%)
    
    Returns:
        float: margin of error
    """
    data = np.array(data)
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    se = sample_std / np.sqrt(n)       # Standard error of the mean

    # Critical t-value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)
    
    # Margin of error
    moe = t_crit * se
    
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Size: {n}")
    print(f"Standard Deviation: {sample_std:.2f}")
    print(f"{int(confidence*100)}% Margin of Error: ±{moe:.2f}")
    
    return moe

# 🔍 Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=50, scale=10, size=40)
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
#Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
#Ans.
def bayes_theorem(prior_disease, sensitivity, false_positive_rate):
    # Calculate the total probability of a positive result
    # P(Positive) = P(Pos|Disease) * P(Disease) + P(Pos|No Disease) * P(No Disease)
    prior_no_disease = 1 - prior_disease
    p_positive = sensitivity * prior_disease + false_positive_rate * prior_no_disease

    # Apply Bayes' Theorem
    posterior = (sensitivity * prior_disease) / p_positive

    print("Bayesian Inference Using Bayes’ Theorem")
    print("---------------------------------------")
    print(f"Prior Probability of Disease: {prior_disease:.2%}")
    print(f"Sensitivity (P(Positive | Disease)): {sensitivity:.2%}")
    print(f"False Positive Rate (P(Positive | No Disease)): {false_positive_rate:.2%}")
    print(f"Total Probability of Positive Test: {p_positive:.2%}")
    print(f"\nPosterior Probability (P(Disease | Positive)): {posterior:.2%}")
    return posterior

# 🔍 Example: Test Parameters
prior_disease = 0.01           # 1% prevalence
sensitivity = 0.99             # 99% true positive
false_positive_rate = 0.05     # 5% false positive

bayes_theorem(prior_disease, sensitivity, false_positive_rate)
```
